### Data Visualization <br>
<p>Data visualization is one of the most powerful tools an analyst has for two main reasons. Firstly, it is unrivalled in it's ability to guide the analysts hand in determining 'what to look at next'. Often a visual is revealing of patterns in the data that are not easily discernable by just looking at dataframes. Secondly, they are an analysts greatest communication tool. Very often professional analysts need to present their results to groups of people responsible for taking action based on what the data says. Visuals can tell your story much better then raw numbers, so let's talk a little bit about how to make effective visuals that make a point.</p>

##### Importing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt # standard
```

### Different Types of Charts <br>
<ul>
    <li><b>Line Chart:</b> Exploring data over time</li>
    <li><b>Bar Chart:</b> Comparing categories of data and tracks changes over time</li>
    <li><b>Pie Chart:</b> Explore parts of a whole, i.e. fractions</li>
    <li><b>Scatter Plot:</b> Similar to line charts, track correlations between two categories</li>
    <li><b>Histogram:</b> Unrelated from bar charts, show distribution of variables</li>
    <li><b>Candlestick Chart:</b> Used a lot in financial sector, i.e. can compare a stock over a period of time</li>
    <li><b>Box Chart:</b> Looks identical to candlestick charts, compares minimum, 1st, medium, 3rd quartiles, and max values</li>
</ul>

<p>Depending on what you need to accomplish in conceptualizing your data, you will go with a specific type of chart to portray your data.</p>

### Subplots <br>
<p>Using the subplots method, you're able to create multiple plots within a single cell.</p>

```python
# first plot
plot_1 = plt.subplot(2,1,1) # 2 rows, 1 column, first panel position in the plot
plt.plot([1600, 1700, 1800, 1900, 2000], [1,1.4,1.6,1.8,2], 'ro-.', label="First Demographic")
plot_1.set_xlim([1550,2050])
plot_1.set_ylim([0,3])

plt.ylabel('Population (billions)')
plt.xlabel('Year')
plt.legend()
# second plot

plot_2 = plt.subplot(2,1,2) # 2 rows , 1 column, second panel position in the plot
plt.plot([1600,1700,1800,1900,2000], [2,2,3,3.1,2.4], 'gx-', label='Second Demographic')
plot_2.set_xlim([1550,2050])
plot_2.set_ylim([1,4])
plt.ylabel('Population (billions)')
plt.xlabel('Year')
plt.legend()

plt.show()
```

### Presentation Worthiness
<p>Before we go into how to use this as a tool for data exploration, let's talk quickly about presentation-worthiness. In the next cell we're going to introduce customization of lines, colors, ticks, titles, fonts, and grids.</p>

```python
fig, ax = plt.subplots(figsize=(20, 10))

plt.plot(x**2, y**4,
        lw=5, # line width
        ls='--', # line style
        color='r',
        marker='*', # point maker
        markersize=15,
        markerfacecolor='b', # blue color
        label='Data'
    )

# TeX language used below, prefaced with the 'r' for raw string, $ starts it and $ ends it
plt.xlabel(r'$x$ Value', fontname='Times New Roman', fontsize=32)
plt.ylabel(r'$y$ Value', fontname = 'Times New Roman', fontsize=32)

# accessing ticks
plt.xticks(fontsize=28)
plt.yticks(fontsize=28)

# adding a title to the chart
plt.title("Quadratic Equation Visual", fontname="Times New Roman", fontsize=40, y=1.05)
# show grid lines
plt.grid(True)
# adjusting the legen
plt.legend(loc='upper left', fontsize=40)
# show chart
plt.tight_layout()
```

<p>It's sometimes nice to obsess a little bit over how your plots look. It's good practice paying attention to detail and it has huge payoff when you end up presenting your findings to other analysts and business parteners. Here's a more complex example of using the matplotlib api:</p>

```python
markers = ['o', 'v', '^', '*', 's']
styles = ['-', '--', '-.', ':', '-']
colors = ['#588C7E', '#F2E394', '#F2AE72', '#D96459', '#8C4646']
```

```python
x = np.linspace(0, 10, 50)

fig, ax = plt.subplots(figsize=(20, 10))

# loop in a range of 5 to create separate lines within the chart
for i in range(5):
    plt.plot(x, x**i,
            lw = 5, # line width
            ls = styles[i], # line style
            color = colors[i],
            marker = markers[i], # point maker
            markersize=15,
            markerfacecolor = colors[i], # blue color
            label = r'$x^{%s}$' % i
        )
    

plt.xlim(0, 10, 50)
plt.ylim(0, 100)

# TeX language used below, prefaced with the 'r' for raw string, $ starts it and $ ends it
plt.xlabel(r'$x$ Value', fontname='Times New Roman', fontsize=32)
plt.ylabel(r'$y$ Value', fontname='Times New Roman', fontsize=32)

# accessing ticks
plt.xticks(fontsize=28)
plt.yticks(fontsize=28)

# adding a title to the chart
plt.title(r'$y = f(x) = x^2$', fontname='Times New Roman', fontsize=40, y=1.02)

# show grid lines
plt.grid(True)

# adjusting the legend
plt.legend(loc='upper left', fontsize=40)

# show chart
plt.tight_layout()
```

### Scatter Plot

```python
import csv
from  itertools import groupby

data_file = 'files/iris.csv'
plt.figure(figsize=(10.5, 7.25)) # similar to variable assignment, but accessing plt method

def read_csv(filename):
    with open(filename, 'r') as f:
        data = list(csv.reader(f))
    return data

flowers = read_csv(data_file)
# print(flowers[0])

colors = {
    'Iris-setosa': '#2B5B84',
    'Iris-versicolor': 'g',
    'Iris-virginica': 'purple'
}

flowers.pop()

for species, group in groupby(flowers, lambda x: x[4]):
    flower_cat = list(group)
    sepal_length = [float(flower[0]) for flower in flower_cat]
    sepal_width = [float(flower[1]) for flower in flower_cat]
    plt.scatter(sepal_length, sepal_width, s=10, c=colors[species], label=species)
    
plt.title('Iris Dataset Sepal Length x Width', fontsize=16)
plt.xlabel('Sepal length (cm)', fontsize=10)
plt.ylabel('Sepal width (cm)', fontsize=10)
plt.legend(loc='best').get_frame().set_color('chocolate')
plt.show()
```

### Box Plot

```python
from itertools import groupby

data_file = 'files/iris.csv'
plt.figure(figsize=(10.5, 7.25))

fig = plt.figure()

def read_csv(filename):
    with open(filename, 'r') as file:
        data = list(csv.reader(file))
    return data

flowers = read_csv(data_file)

# Find pedal length inside of a box plot

petal_length = []

for species, group in groupby(flowers, lambda x:x[4] ):
#     print(species)
#     print(group)
    petal_length.append([float(petal[2]) for petal in group])
print(petal_length)

plt.boxplot(petal_length)
plt.axis([0,4,0,10]) # xmin, xmax, ymin, ymax
plt.xticks([1, 2, 3], ['Setosa', 'Versicolor', 'Virginica'])
plt.title('Box Plot with Iris Dataset', fontsize=16, loc='right')
plt.xlabel('Flower Variety', fontsize=10)
plt.ylabel('Petal Length (cm)', fontsize=10)

plt.show()

fig.savefig('files/iris_petal_data.png') # Saving as a png in the local files folder
```

### Histograms <br>
<p>While line plots are great for visualizing trends in time series data, histograms are the king of visualizing distributions. Often, the distribution of a variable is what you're interested in, and a visualization provides a lot more information than a group of summary statistics.</p>

```python
fig, ax = plt.subplots(figsize=(16, 9)) #Subplots makes a call in the background to create a plot area


# generate random data
data = np.random.uniform(0, 10, 100)

# create a histogram plot
plt.hist(data, bins=50, label="Data", color = 'r', alpha = 0.75)
plt.xticks(fontsize=30)
plt.yticks(fontsize=30)
plt.xlabel('Value', fontname='Times New Roman', fontsize=35)
plt.ylabel('Frequency', fontname='Times New Roman', fontsize = 35)
plt.legend(loc='upper left', fontsize = 30)

plt.show()
```

<p>Let's take a look at a mystery/randomly generated DataFrame, and plot the results</p>

```python
import pandas as pd
import numpy as np
def make_mystery_dataframe():
    customer_id = np.random.randint(100000,120000,10000)
    
    
    states = {1:'California', 0:'New York'}
    statelist = []
    agelist = []
    for i in range(10000):
        ind = np.random.randint(0, 2)
        if ind == 0:
            statelist.append(states[ind])
            agelist.append(np.random.normal(loc=30, scale=10))
        elif ind == 1:
            statelist.append(states[ind])
            agelist.append(np.random.normal(loc=40, scale=20))
    
    mystery_df = pd.DataFrame.from_dict({'customer_id': customer_id,
                                         'ages':agelist,
                                         'state':statelist})
    
    return mystery_df
            

mystery_df = make_mystery_dataframe()
mystery_df.head()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 10))

# first subplot
ax1.hist(mystery_df[mystery_df['state'] == 'California']['ages'], bins=50, label="Ages in Cali")
ax1.set_xlabel('Value', fontname="Times New Roman", fontsize=35)
ax1.set_ylabel('Frequency', fontname = 'Times New Roman', fontsize = 35)
ax1.tick_params(axis='both', labelsize=30)
ax1.legend(loc = 'upper left', fontsize = 30)
# second subplot
ax2.hist(mystery_df[mystery_df['state'] == 'New York']['ages'], bins=50, label="Ages in NY")
ax2.set_xlabel('Value', fontname="Times New Roman", fontsize=35)
ax2.set_ylabel('Frequency', fontname = 'Times New Roman', fontsize = 35)
ax2.tick_params(axis='both', labelsize=30)
ax2.legend(loc = 'upper left', fontsize = 30)

plt.xticks(fontsize=30)
```

### 3-D Plotting <br>
<p>Without going into too much detail, here's a quick example of the type of 3-D plotting that's possible with matplotlib (example stolen from the official docs):</p>

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.tri as mtri

fig = plt.figure(figsize=(20,10))

#============
# First plot
#============

# Make a mesh in the space of parameterisation variables u and v
u = np.linspace(0, 2.0 * np.pi, endpoint=True, num=50)
v = np.linspace(-0.5, 0.5, endpoint=True, num=10)
u, v = np.meshgrid(u, v)
u, v = u.flatten(), v.flatten()

# This is the Mobius mapping, taking a u, v pair and returning an x, y, z
# triple
x = (1 + 0.5 * v * np.cos(u / 2.0)) * np.cos(u)
y = (1 + 0.5 * v * np.cos(u / 2.0)) * np.sin(u)
z = 0.5 * v * np.sin(u / 2.0)

# Triangulate parameter space to determine the triangles
tri = mtri.Triangulation(u, v)

# Plot the surface.  The triangles in parameter space determine which x, y, z
# points are connected by an edge.
ax = fig.add_subplot(1, 2, 1, projection='3d')
ax.plot_trisurf(x, y, z, triangles=tri.triangles, cmap=plt.cm.Spectral)
ax.set_zlim(-1, 1)


#============
# Second plot
#============

# Make parameter spaces radii and angles.
n_angles = 36
n_radii = 8
min_radius = 0.25
radii = np.linspace(min_radius, 0.95, n_radii)

angles = np.linspace(0, 2*np.pi, n_angles, endpoint=False)
angles = np.repeat(angles[..., np.newaxis], n_radii, axis=1)
angles[:, 1::2] += np.pi/n_angles

# Map radius, angle pairs to x, y, z points.
x = (radii*np.cos(angles)).flatten()
y = (radii*np.sin(angles)).flatten()
z = (np.cos(radii)*np.cos(angles*3.0)).flatten()

# Create the Triangulation; no triangles so Delaunay triangulation created.
triang = mtri.Triangulation(x, y)

# Mask off unwanted triangles.
xmid = x[triang.triangles].mean(axis=1)
ymid = y[triang.triangles].mean(axis=1)
mask = np.where(xmid**2 + ymid**2 < min_radius**2, 1, 0)
triang.set_mask(mask)

# Plot the surface.
ax = fig.add_subplot(1, 2, 2, projection='3d')
ax.plot_trisurf(triang, z, cmap=plt.cm.CMRmap)


plt.show()
```

### Flattening Multidimensional Data

<p>Generally in data science you want to avoid 3D plotting wherever possible - not because the information you want to convey is not contained within the result, sometimes it is simply because your point is easier to express by denoting one of the axes using color or some other means. For instance, imagine you are asked to investigate the effects of driver age and car engine size on number of traffic violiations. That is, you are attempting to describe a function - the number of traffic violations - of two variables, age and engine size. You could make a 3D scatterplot, but let's see if we can come up with something better:</p>

```python
def random_line(m, b, sigma, size=150):
    '''create some fake data about drivers!'''
    
    
    xdata = np.linspace(18,100,size)
    # Generate normally distributed random error ~ N(0, sigma**2)
    #errors = norm.rvs(loc=0, scale=sigma, size=size)
    errors = np.random.uniform(20,100,size) + np.random.uniform(20,100,size) + np.random.uniform(20,100,size)
    ydata = (m * xdata + b + errors) / 20
    
    num_accidents = (np.round(np.random.uniform(0,1,size) + (10-xdata)+ydata) / 10 + 10) + (ydata/5)
    return xdata, ydata-7, num_accidents
```

```python
ages, engine_sizes, num_accidents = random_line(1.35, 2, 3)
```

```python
fig, ax = plt.subplots(figsize=(12, 7))
plt.scatter(ages,engine_sizes, c = num_accidents)
plt.xlabel('Driver Age', fontsize = 19)
plt.ylabel('Car Engine Size', fontsize = 19)
plt.colorbar(label = 'Traffic Violations')
```
